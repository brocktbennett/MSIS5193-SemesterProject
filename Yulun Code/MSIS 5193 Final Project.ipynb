```python
## Here, I tried reservoir sampling to randomly sample 50k rows from a total of 5.4 million rows
    # while keepin the class ratio of the target variable.
        # stratified sampling is another sampling method
```

```python
# perform reservoir sampling in df_sub_no_nan_cat 
    # while keeping the proportions of the two classes of the target column ("LAW_CAT_CD_binary_tgt")

import pandas as pd
import numpy as np

def reservoir_sampling_binary(df, binary_column, k):
    """
    Perform reservoir sampling on a DataFrame based on a binary column.
    
    Parameters:
    df (DataFrame): The DataFrame to sample from.
    binary_column (str): The name of the binary column.
    k (int): Total number of samples to select.
    
    Returns:
    DataFrame: A new DataFrame containing k samples from df.
    """
    # Calculate the proportions of the two classes
    prop_0 = sum(df[binary_column] == 0) / len(df)
    prop_1 = 1 - prop_0

    # Calculate the number of samples for each class
    k_0 = round(k * prop_0)
    k_1 = k - k_0

    # Sample k_0 rows from class 0 and k_1 rows from class 1
    sampled_df = pd.concat([
        df[df[binary_column] == 0].sample(n=k_0, replace=True),
        df[df[binary_column] == 1].sample(n=k_1, replace=True)
    ])

    return sampled_df
```

```python
# (1) perform associations rule mining

# metrics to consider (https://dataaspirant.com/association-rule-analysis/):
    # Support: This measures the frequency of occurrence of an itemset (i.e., a set of items) in the dataset. Rules with high support are more significant as they occur more frequently in the dataset.

    # Confidence: This measures the conditional probability of the occurrence of the consequent given the antecedent. Rules with high confidence are more reliable, as they have a higher probability of being true.

    # Lift: This measures the ratio of the observed support to that expected if the antecedent and the consequent were independent. Rules with high lift indicate a strong association between the antecedent and consequent, as they occur together more frequently than expected by chance.

        # Each of these metrics provides a different perspective on the rules, and they are often used together to provide a more comprehensive evaluation. For example, you might first use support to identify frequently occurring itemsets, then use confidence to find rules that have a high probability of being true, and finally use lift to find rules that represent strong associations2. It’s always a good idea to experiment with different metrics and see which ones provide the most useful and interpretable results for your specific use case.
```

```python

# prepare dataset (since associations rule mining examines unique items, the columns should not be coded using integers, i.e., each column should show the corresponding values, such as seasons, boros, etc)
    # make sure the dataset contains no missing value
    # make sure the item names (i.e., column values) are not numerically encoded

## map the unique values of columns by string values
    # ['holiday_ind'] - 0: non-holiday, 1: holiday
    # ['season_ind'] - 'Spring':1, 'Summer':2, 'Fall':3, 'Winter':4
    # ['LAW_CAT_CD_binary_tgt'] - 0: Misdemeanor & Violation, 1: Felony
    # ['ARREST_BORO_coded'] - 'Brooklyn':0, 'Manhattan':1, 'Bronx':2, 'Queens':3, 'Staten Island':4
    # ['JURISDICTION_CODE_coded'] - 0: NYPD_Patrol, 1: NYPD_Transit, 2: NYPD_Housing, 3: Non-NYPD
    # ['AGE_GROUP_coded'] - '<18':0, '18-24':1, '25-44':2, '45-64':3, '65+':4
    # ['PERP_SEX_coded'] - 'Male':0, 'Female':1
    # ['PERP_RACE_coded'] - 'BLACK':0, 'WHITE HISPANIC':1, 'WHITE':2, 'BLACK HISPANIC':3, 'ASIAN/PACIFIC ISLANDER':4, 'AMERICAN INDIAN/ALASKAN NATIVE':5, 'UNKNOWN & OTHER':6

sampled_df_cat_val_0 = sampled_df_cat.copy()

sampled_df_cat_val_0['holiday_ind'] = sampled_df_cat_val_0['holiday_ind'].map({0: 'non-holiday', 1: 'holiday'})

sampled_df_cat_val_0['season_ind'] = sampled_df_cat_val_0['season_ind'].map({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})

sampled_df_cat_val_0['LAW_CAT_CD_binary_tgt'] = sampled_df_cat_val_0['LAW_CAT_CD_binary_tgt'].map({0: 'Misdemeanor & Violation', 1: 'Felony'})

sampled_df_cat_val_0['ARREST_BORO_coded'] = sampled_df_cat_val_0['ARREST_BORO_coded'].map({0: 'Brooklyn', 1: 'Manhattan', 2: 'Bronx', 3: 'Queens', 4: 'Staten Island'})

sampled_df_cat_val_0['JURISDICTION_CODE_coded'] = sampled_df_cat_val_0['JURISDICTION_CODE_coded'].map({0: 'NYPD_Patrol', 1: 'NYPD_Transit', 2: 'NYPD_Housing', 3: 'Non-NYPD'})

sampled_df_cat_val_0['AGE_GROUP_coded'] = sampled_df_cat_val_0['AGE_GROUP_coded'].map({0: 'Younger Than 18', 1: '18 to 24 Years Old', 2: '25 to 44 Years Old', 3: '45 to 64 Years Old', 4: 'Older Than 65'})

sampled_df_cat_val_0['PERP_SEX_coded'] = sampled_df_cat_val_0['PERP_SEX_coded'].map({0: 'Male', 1: 'Female'})

sampled_df_cat_val_0['PERP_RACE_coded'] = sampled_df_cat_val_0['PERP_RACE_coded'].map({0: 'Race_Black', 1: 'Race_White Hispanic', 2: 'Race_White', 3: 'Race_Black Hispanic', 4: 'Race_Asian/Pacific Islander', 5: 'Race_American Indian/Alaskan Native', 6: 'Race_Unknown & Other Race'})

sampled_df_cat_val_0.head()
```

```python
# perform associations rule mining (the code below is based on Assignment 6)

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# initialize an empty set to get all unique items
all_items = set()

# find all the unique products
for index, row in sampled_df_cat_val.iterrows():
    # update() method updates the current set, by adding items from row
    all_items.update(row)  # https://www.w3schools.com/python/ref_set_update.asp#:~:text=The%20update()%20method%20updates,present%20in%20the%20updated%20set
print("Count of items: ", len(all_items))
print("Items: ", all_items)
```

```python
# One Hot Encoding
encoded_vals = list()

# iterate through dataframe to convert each cell into 1 - has product or 0 - no product
for index, row in sampled_df_cat_val.iterrows():
    # set all cells in a row as 0
    rowset = {item: 0 for item in all_items}
    # iterate through the row, if the item is no 0, set the value to be 1
    rowset.update({item: 1 for item in row if item != 0})
    # add the rowset to the list
    encoded_vals.append(rowset)

# convert the list into a new dataframe
encoded_vals_df = pd.DataFrame(encoded_vals)

encoded_vals_df
```

```python
# generate freq item sets and set support threshold min_support
freq_items = apriori(encoded_vals_df, min_support=0.1, use_colnames=True)
pd.set_option('display.max_columns', None)

# examine the frequent items
freq_items
```

```python
# get the rules by setting confidence threshold
rules = association_rules(freq_items, metric="confidence", min_threshold=0.1)
pd.set_option('display.max_columns', None)

# examine the rules
rules
```

```python
# get the columns 'antecedents', 'consequents', 'support', 'confidence', and 'lift' from the rules
rules_sub = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
pd.set_option('display.max_columns', None)

# examine rules_sub
rules_sub
```

```python
# to determine k, two of the most commonly used methods are the Elbow Method and the Silhouette Method:

# Elbow Method: 
    # This method involves running the K-Means algorithm for a range of values of k (for example, from 1 to 10), and for each value of k, calculating the Within-Cluster-Sum of Squared Errors (WCSS). WCSS is the sum of the squared distance between each point and its centroid. You then plot the WCSS values against the number of clusters. The optimal number of clusters is at the "elbow" point in the plot, where the rate of decrease in WCSS starts to diminish. And because the interpretation of the "elbow" point can be subjective, potentially making the Elbow Method unreliable.

# Silhouette Method: 
    # This method measures how similar a point is to its own cluster (cohesion) compared to other clusters (separation). The silhouette value ranges between -1 and +1, with a high value indicating that the point is well-matched to its own cluster and poorly matched to neighboring clusters. If the silhouette score for various values of k is plotted, the most optimal number of clusters will be the one with the highest average silhouette score. 
```

```python
# determine "k" using the Silhouette Method
    # drop rows with missing values (required for k-means clustering)
    # DO NOT INCLUDE TARGET VARIABLE!

import numpy as np
import matplotlib.pyplot as plt
#from sklearn.cluster import KMeans  # for continous variables
#from sklearn.cluster import MiniBatchKMeans # for continuous variables
    # MiniBatchKMeans uses mini-batches to reduce computation time while producing results that are slightly worse than the regular K-Means; https://scikit-learn.org/stable/modules/clustering.html#mini-batch-kmeans
from kmodes.kmodes import KModes
from sklearn.metrics import silhouette_score

# Prepare data - use the dataframe without missing values (df_sub_no_nan) & drop unwanted variables (including target) for k-means clustering; in addition, holiday_ind is extremely unbalanced - drop it from clustering.

df_clustering = sampled_df_cat.drop(['LAW_CAT_CD_binary_tgt', 'holiday_ind'], axis=1)

# in addition, PERP_SEX_coded and JURISDICTION_CODE_coded shows extreme unbalanced class ratio
    # however, considering that the clustering results will be used for modeling, they are not dropped from clustering analysis

# df_clustering = sampled_df_cat.drop(['LAW_CAT_CD_binary_tgt', 'holiday_ind', 'PERP_SEX_coded', 'JURISDICTION_CODE_coded'], axis=1)

df_clustering.info()
```

```python
# determine the optimal number of clusters ("k") using Silhouette Method

# Define the range of k values to test
k_range = range(2, 11)

# Initialize lists to store the results
silhouette_scores = []

# Placeholder for the best silhouette score and the corresponding number of clusters
best_n_clusters = None
best_silhouette_score = -1

# Loop over the range of k values
for k in k_range:
    # Initialize the k-means model
    # kmeans = KMeans(n_clusters=k, random_state=42)
    # kmeans = MiniBatchKMeans(n_clusters=k, batch_size=1000, random_state=42)
    kmode = KModes(n_clusters=k, init = "Cao", n_init = 1, verbose=1)

    # Fit the model to the data to find the optimal number of clusters
    kmode.fit_predict(df_clustering)

    # Get the cluster labels for each data point
    labels = kmode.labels_

    # Calculate the silhouette score for each data point
    scores = silhouette_score(df_clustering, labels)

    # Calculate the average silhouette score for this value of k
    avg_score = np.mean(scores)

    # Append the results to the lists
    silhouette_scores.append(avg_score)

    # Check if this silhouette score is the best so far
    if avg_score > best_silhouette_score:
        best_silhouette_score = avg_score
        best_n_clusters = k

# Plot the results
plt.plot(k_range, silhouette_scores, 'bo-')
plt.xticks(range(2, 11))
plt.xlabel('Number of Clusters')
plt.ylabel('Average Silhouette Score')
plt.title('Silhouette Method for Optimal Clustering')
plt.show()
```

```python
# The highest average silhouette score indicates the optimal number of clusters.
    # in this case, the optimal number of clusters is 2
```

```python
# Add the cluster labels to the DataFrame

# Now that we have the best number of clusters, perform k-modes clustering again with this number of clusters
km_best = KModes(n_clusters=best_n_clusters, init='Huang', n_init=5, verbose=0)
best_clusters = km_best.fit_predict(df_clustering)

# Add the cluster labels to the DataFrame
df_clustering_clusters = df_clustering.assign(cluster_label=best_clusters)

df_clustering_clusters.head()
```

```python
# build a lightGBM model to predict if a felony will happen or not ("LAW_CAT_CD_binary_tgt"):
    # use the dataset which has applied with Catboost Encoding

    # tune hyperparameters using Random Search
        # add cross validation to reduce overfitting
        # add weights to address unbalanced target by reduce model biasness towards the majority class
    # evaluate model performance
    # then evaluate feature importance using both split and gain
    # then retrain the model using the most important variables & evaluate model performance

import numpy as np
from sklearn.model_selection import RandomizedSearchCV
import lightgbm as lgb
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, accuracy_score

sampled_df_cat_val.columns.tolist()
```

```python
# train-test split (70-30) 

sampled_df_cat_val_1 = sampled_df_cat_val.copy()
    # make sure to perform numerical encoding to target variable
sampled_df_cat_val_1['LAW_CAT_CD_binary_tgt'] = sampled_df_cat_val_1['LAW_CAT_CD_binary_tgt'].map({'Misdemeanor & Violation': 0, 'Felony': 1})

# define X and y
X_sampled = sampled_df_cat_val_1[['holiday_ind', 'season_ind', 'ARREST_BORO_coded', 'JURISDICTION_CODE_coded', 'AGE_GROUP_coded', 'PERP_SEX_coded', 'PERP_RACE_coded']]

y_sampled = sampled_df_cat_val_1['LAW_CAT_CD_binary_tgt']

# split the data into 70% training and 30% validation sets; 
    # set stratify=y to keep the target class proportion.
from sklearn.model_selection import train_test_split
X_train_sampled, X_test_sampled, y_train_sampled, y_test_sampled = train_test_split(X_sampled, y_sampled, test_size=0.3, random_state=42, stratify=y_sampled)

print(len(X_train_sampled))
print(len(X_test_sampled))
print(len(y_train_sampled))
print(len(y_test_sampled))
```

```python
# perform CatBoost Encoding, which encodes the categorical variables based on target variable classes

# Create an instance of CatBoostEncoder
import category_encoders as ce 
cbe_encoder = ce.cat_boost.CatBoostEncoder(random_state=1321)

# Fit encoder and transform the features
cbe_encoder.fit(X_sampled, y_sampled) 

# Encode categorical variables in both training and validation data
X_train_CatBoostEncoded = cbe_encoder.transform(X_train_sampled)
X_test_CatBoostEncoded = cbe_encoder.transform(X_test_sampled)

# examine the encoded dataframe - train
X_train_CatBoostEncoded.info()
X_train_CatBoostEncoded.head()
```

```python
# examine the encoded dataframe - test
X_test_CatBoostEncoded.info()
X_test_CatBoostEncoded.head()
```

```python
# Fit LightGBM Classifier by tuning hyperparameters using Random Search Cross Validation (RandomSearchCV) 
    # add early stopping to reduce overfitting
#https://lightgbm.readthedocs.io/en/stable/Parameters.html
#https://www.geeksforgeeks.org/regularization-by-early-stopping/

# Define the hyperparameter configuration space
param_dist_lgbm = {"max_depth": [5, 10, 20],
                  "min_samples_leaf": [5, 10, 20, 50, 100],
                  'num_leaves': [5, 20, 31],
                  'learning_rate': [0.05, 0.1, 0.2],
                  'colsample_bytree': [0.3, 0.5, 0.7],
                  'subsample': [0.5, 0.7, 0.9],
                  'num_iterations': [50, 100, 150],
                  'lambda_l1': [0, 0.1, 1],
                  'lambda_l2': [0, 0.1, 1]
}

# Instantiate a lightGBM classifier
lgbm_sampled = lgb.LGBMClassifier(objective='binary', importance_type='gain')

# Instantiate the RandomizedSearchCV object
lgbm_sampled_cv = RandomizedSearchCV(lgbm_sampled, param_dist_lgbm, scoring='roc_auc', cv=5)

# Fit it to the training data
lgbm_sampled_cv.fit(X_train_CatBoostEncoded, y_train_sampled)

# Print the tuned parameters and score
print("Tuned LightGBM Parameters: {}".format(lgbm_sampled_cv.best_params_))
print("Best training score is {}".format(lgbm_sampled_cv.best_score_))
```

```python
# In the lightGBM model built above, the tuned hyperparameters include 'subsample' (0.7), 'colsample_bytree' (0.5), 'learning_rate'(0.05), 'max_depth' (5), 'min_samples_leaf' (100), 'num_iterations' (150), 'num_leaves' (20), 'lambda_l1' (0), 'lambda_l2' (0.1).
# In specific,  'colsample_bytree' indicates the number of rows that is used to train a tree (i.e., learner) without re-sampling; for 'subsample', it has a similar meaning with 'colsample_bytree' but without re-sampling; 'learning_rate' controls the pace the model learns the data (smaller value means lower learning pace); 'max_depth' means the maximum depth from the root node to the leaf node of each tree in the forest, which controls the overall complexity (i.e., underfitting vs. overfitting) of each tree. Similarly, 'num_leaves' means the maximum number of leaves in each learner; 'min_samples_leaf' means the minimum number of data in each leaf; 'num_iterations' means the number of boosting iterations; 'lambda_l1' and 'lambda_l2' means L1 regularization and L2 regularization, respectively.
```

```python
# calculate predicted probability using X_test_CatBoostEncoded
y_test_sampled_pred_prob_lgbm = lgbm_sampled_cv.predict_proba(X_test_CatBoostEncoded)[:,1]

# Calculate AUC score using predicted probabilities on Test Set
auc_test_sampled_lgbm = roc_auc_score(y_test_sampled, y_test_sampled_pred_prob_lgbm)

print('AUC of lightGBM classifier using sampled data: %.3f' % auc_test_sampled_lgbm)
```

```python
# to prepare the data for Catboost Encoding: map the unique values of columns by string values
    # ['holiday_ind'] - 0: non-holiday, 1: holiday
    # ['season_ind'] - 'Spring':1, 'Summer':2, 'Fall':3, 'Winter':4
    # ['LAW_CAT_CD_binary_tgt'] - 0: Misdemeanor & Violation, 1: Felony
    # ['ARREST_BORO_coded'] - 'Brooklyn':0, 'Manhattan':1, 'Bronx':2, 'Queens':3, 'Staten Island':4
    # ['JURISDICTION_CODE_coded'] - 0: NYPD_Patrol, 1: NYPD_Transit, 2: NYPD_Housing, 3: Non-NYPD
    # ['AGE_GROUP_coded'] - '<18':0, '18-24':1, '25-44':2, '45-64':3, '65+':4
    # ['PERP_SEX_coded'] - 'Male':0, 'Female':1
    # ['PERP_RACE_coded'] - 'BLACK':0, 'WHITE HISPANIC':1, 'WHITE':2, 'BLACK HISPANIC':3, 'ASIAN/PACIFIC ISLANDER':4, 'AMERICAN INDIAN/ALASKAN NATIVE':5, 'UNKNOWN & OTHER':6
    # ['LAW_CAT_CD_binary_tgt'] - 0: Misdemeanor & Violation, 1: Felony
    # ['cluster_label'] - 1: Cluster_0, 1: Cluster_1

df_clustering_tgt_val = df_clustering_tgt.copy()

df_clustering_tgt_val['cluster_label'] = df_clustering_tgt_val['cluster_label'].map({0: 'Cluster_0', 1: 'Cluster_1'})

df_clustering_tgt_val['season_ind'] = df_clustering_tgt_val['season_ind'].map({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})

# df_clustering_tgt_val['LAW_CAT_CD_binary_tgt'] = df_clustering_tgt_val['LAW_CAT_CD_binary_tgt'].map({0: 'Misdemeanor & Violation', 1: 'Felony'})

df_clustering_tgt_val['ARREST_BORO_coded'] = df_clustering_tgt_val['ARREST_BORO_coded'].map({0: 'Brooklyn', 1: 'Manhattan', 2: 'Bronx', 3: 'Queens', 4: 'Staten Island'})

df_clustering_tgt_val['JURISDICTION_CODE_coded'] = df_clustering_tgt_val['JURISDICTION_CODE_coded'].map({0: 'NYPD_Patrol', 1: 'NYPD_Transit', 2: 'NYPD_Housing', 3: 'Non-NYPD'})

df_clustering_tgt_val['AGE_GROUP_coded'] = df_clustering_tgt_val['AGE_GROUP_coded'].map({0: 'Younger Than 18', 1: '18 to 24 Years Old', 2: '25 to 44 Years Old', 3: '45 to 64 Years Old', 4: 'Older Than 65'})

df_clustering_tgt_val['PERP_SEX_coded'] = df_clustering_tgt_val['PERP_SEX_coded'].map({0: 'Male', 1: 'Female'})

df_clustering_tgt_val['PERP_RACE_coded'] = df_clustering_tgt_val['PERP_RACE_coded'].map({0: 'Race_Black', 1: 'Race_White Hispanic', 2: 'Race_White', 3: 'Race_Black Hispanic', 4: 'Race_Asian/Pacific Islander', 5: 'Race_American Indian/Alaskan Native', 6: 'Race_Unknown & Other Race'})

df_clustering_tgt_val.head()
```

```python
# train-test split (70-30) 

# define X and y
X_sampled_k = df_clustering_tgt_val[['cluster_label', 'season_ind', 'ARREST_BORO_coded', 'JURISDICTION_CODE_coded', 'AGE_GROUP_coded', 'PERP_SEX_coded', 'PERP_RACE_coded']]

y_sampled_k = df_clustering_tgt_val['LAW_CAT_CD_binary_tgt']

# split the data into 70% training and 30% validation sets; 
    # set stratify=y to keep the target class proportion.
from sklearn.model_selection import train_test_split
X_train_sampled_k, X_test_sampled_k, y_train_sampled_k, y_test_sampled_k = train_test_split(X_sampled_k, y_sampled_k, test_size=0.3, random_state=42, stratify=y_sampled_k)
```

```python
# perform CatBoost Encoding, which encodes the categorical variables based on target variable classes

# Create an instance of CatBoostEncoder
import category_encoders as ce 
cbe_encoder = ce.cat_boost.CatBoostEncoder(random_state=1321)

# Fit encoder and transform the features
cbe_encoder.fit(X_sampled_k, y_sampled_k) 

# Encode categorical variables in both training and validation data
X_train_CatBoostEncoded_k = cbe_encoder.transform(X_train_sampled_k)
X_test_CatBoostEncoded_k = cbe_encoder.transform(X_test_sampled_k)
```
